```python
f = pd.read_csv('/content/drive/Shareddrives/캡스톤summa/데이터팀/트렌드.csv')
f.columns = ['tags', 'date', 'likes'] #칼럼 지정
```

```python
f = f.drop_duplicates('tags', keep='last')  #tags 리스트 중복 제거
display(f)
```

```python
f['likes'] = f['likes'].str.replace(pat=',', repl = r'') #','문자 제거
f['likes'] = f['likes'].astype(int) #int로 변환
```

```python
idx = f[f['likes'] < 10].index  #좋아요 수 10이하 제거
f = f.drop(idx)
display(f)
```
